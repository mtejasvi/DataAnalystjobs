# Checkpoint Three: Cleaning Data

```python
import pandas as pd
import numpy as np
df = pd.read_csv('DataAnalyst.csv')
df
```

# Missing Data


Test your dataset for missing data and handle it as needed. Make notes in the form of code comments as to your thought process.

```python
df.isna().sum()
# There is only 1 missing data shown in the company column.
# During the EDA there more missing values that were filled with values like -1,-1.0,'-1','1'. 
# If we replace them with null we can find the missing vales from all the columns in the data.
```

```python
df.replace(-1, np.nan, inplace=True)
df.replace(-1.0, np.nan, inplace=True)
df.replace('-1', np.nan, inplace=True)
df.replace('1', np.nan, inplace=True)

df
```

```python
# The following shows the sum of missing values
missing = df.isna().sum()
missing

#Converting the missing values as percentage using 'mean'
missing_pct = round(df.isna().mean()*100,0)
missing_pct

# Easy Apply and competitors columns has the highest missing values.
# In the business issue point of view we may not need those two columns.
# But we need the details of columns 'Rating','Sector' for our analysis.
# Lets try replacing them with a value which will not affect our analysis.
```

```python
df['Rating'].value_counts()
```

```python
# I am replacing the 'NaN' values with 0 as it will not affect the calculation.
# I tried replacing it with 'median' values but it changed the entired rating value counts.
# med = my_data['Rating'].median()
# med
# my_data['Rating'] = my_data['Rating'].fillna(med)
# my_data['Rating'].value_counts()
df['Rating'] = df['Rating'].fillna(0)
df['Rating'].value_counts()
```

```python
# I replaced 'NaN' value with string 'Unknown'. 
# If the missing values are dropped the corresponding rows and their values will be dropped which may affect the result.
df['Sector'] = df['Sector'].fillna('Unknown')
df['Sector'].isnull().sum()
df['Sector'].value_counts()
```

Unnecessary Data

```python
num_rows = len(df)

for col in df.columns:
    cnts = df[col].value_counts(dropna=False)
    top_pct = (cnts/num_rows).iloc[0]
    
    if top_pct > 0.999:
        print('{0}: {1:.2f}%'.format(col, top_pct*100))
        print(cnts)
        print()
        
# Checked for columns that has same repetitive values.There are no such columns.
```

```python
data_copy = df.copy
data_copy
data_copy = df.copy()

# drop unwanted columns
new_data = data_copy.drop(['Unnamed: 0', 'Company Name', 'Location','Headquarters', 'Size', 'Founded', 'Type of ownership', 'Revenue', 'Competitors', 'Easy Apply'], axis=1)
new_data
```

# Inconsistent Data

```python
new_data['Salary Max'] = new_data['Salary Estimate'].str[6:10].str.replace('K', '').str.lstrip('$').str.strip().astype('float')
new_data['Salary Min'] = new_data['Salary Estimate'].str[:3].str.replace('K', '').str.lstrip('$').str.strip().astype('float')
new_data[['Salary Estimate','Salary Max','Salary Min']]
```
